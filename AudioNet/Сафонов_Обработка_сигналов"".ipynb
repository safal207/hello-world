Задание 1:

Переписать функцию Sine , таким образом, чтобы на выходе получать сигнал косинуса вместо синуса (np.cos)


















```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def Cos(A, f, phi, fs, t, N):
    '''
    Inputs:
    A  = Амплитуда сигнала
    f  = Частота сигнала  Hz
    phi  = Начальная фаза синусоиды "Радианы"
    fs = Частота семплирования Hz
    t  = Число вместо 1, для лучшей визуализации графиков
    N  = Длина сигнала

    Output:
    x (numpy array)
    
    '''
    T = t/fs   # Период семплирования
    n = np.arange(N) # Кол-во семплов
    x = A*np.cos(2*np.pi*f*n*T + phi)
    return x

A = 1.0     
f = 10.0    
phi = 1.0   
fs = 50.0   
t = 0.1     

plt.plot(Cos(A, f, phi, fs, t, 100))
```

Задание 2:

Модифицировать функцию Sine таким образом чтобы можно было возвращать составной сигнал - состоящий из комбинаций sin и cos, каждая отдельная гармоническая функция в составном сигнале должна принимать свои параметры:

 A  = Амплитуда сигнала
 
 f  = Частота сигнала  Hz

 phi  = Начальная фаза синусоиды "Радианы"

 Сделайте вариант с 2-мя и 3-мя гармоническими сигналами

```
def Cos_sin( **kwargs):
    '''
    Inputs:
    A  = Амплитуда сигнала
    f  = Частота сигнала  Hz
    phi  = Начальная фаза синусоиды "Радианы"
    fs = Частота семплирования Hz
    t  = Число вместо 1, для лучшей визуализации графиков
    N  = Длина сигнала

    Outputs:
    x1, x2 (numpy array)

    '''
    t = kwargs['t']
    fs = kwargs['fs']
    N = kwargs['N']
    T = t/fs         # Период семплирования
    n = np.arange(N) # Кол-во семплов

    # Вариант с двумя гармоническмим сигналами
    x1 = kwargs['A1']*np.cos(2*np.pi*kwargs['f1']*n*T + kwargs['phi1']) +\
         kwargs['A2']*np.sin(2*np.pi*kwargs['f2']*n*T + kwargs['phi2'])

    # Вариант с тремя гармоническмим сигналами
    x2 = kwargs['A1']*np.cos(2*np.pi*kwargs['f1']*n*T + kwargs['phi1']) +\
         kwargs['A2']*np.sin(2*np.pi*kwargs['f2']*n*T + kwargs['phi2']) +\
         kwargs['A3']*np.cos(2*np.pi*kwargs['f3']*n*T + kwargs['phi3'])
    return x1, x2

# Составляем функцию. На выходе будут 2 сигнала.
sig1, sig2 = Cos_sin(fs=50., t=0.1, N=100, A1=1, A2=2, A3=3, f1=10, f2=5, f3=10, phi1=1, phi2=1, phi3=1)

# Строим графики
plt.plot(sig1)
plt.title('Cos + Sin')
plt.grid(True, which='both')
plt.show()

plt.plot(sig2)
plt.title('Cos + Sin + Cos')
plt.grid(True, which='both')
plt.show()
```

```
import matplotlib.pyplot as plt

# Задаем количество сэмплов
n = 40

# Задаем вектор времени
t = np.linspace(0, 1, n, endpoint=True)

# Функция wive
x = np.sin(3*np.pi*t) + np.sin(5*np.pi*t)

# Функция выбора графиков
def plt_sel(s, *args, **kwargs):
    if s == 0:
        return plt.plot(*args)
    if s == 1:
        return plt.stem(*args, **kwargs)
    if s == 2:
        return plt.step(*args)

# Подписи для графиков
t_titles = ['Аналоговый', 'Дискретный', 'Квантованный']

# Строим графики
fig = plt.figure(figsize=(16, 4), dpi=80)
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.title(t_titles[i]) 
    plt_sel(i, t, x, use_line_collection=True)
    plt.xlim([0, 1])
    plt.yticks(np.linspace(np.floor(np.min(x)), np.ceil(np.max(x)), 9))
    plt.grid(True)
    
    # Добавляем отображение оси x
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
plt.tight_layout()
```

```
from scipy.fftpack import fft
import librosa
import librosa.display

# Загружаем пример аудио файла
filename = librosa.util.example_audio_file()
y, sr = librosa.load(filename)
```

Найдите модуль и аргумент спектра для следующего гармонического сигнала: синус (np.sin) С Частотой 20Гц Амплитудой 3 И фазой 1

```
'''
   Модуль спектра называют двусторонним спектром амплитуд(амплитудный спектр)
   или АЧХ - амплитудно-частотной характеристикой сигнала, а аргумент спектра
   - двусторонним спектром фаз(фазовый спектр) или ФЧХ – фазово-частотной
   характеристикой.

'''
# Зададим количество отсчетов
N = 10000 

# Зададим вектор времени
t = np.linspace(0, 1, N)

# Зададим Амплитуды и Частоты
f = 20
A =  3

# Цифровой сигнал
x = A * np.sin(2*np.pi*f*t+1) 

# Создаем 3 subplots
fig, ax = plt.subplots(1, 3, figsize=(30,10))
ax[0].plot(t, x)
ax[0].set_title('Сигнал', fontdict={'fontsize': 20})

# Выводим модуль спектра
a_f = np.fft.rfft(x)*2/N
ax[1].stem(abs(a_f), use_line_collection=True)
ax[1].set_xlim(0, 100)
ax[1].set_title('Модуль спектра', fontdict={'fontsize': 20})

# Выводим аргумент спектра
ax[2].plot(np.angle(a_f))
ax[2].set_xlim(0,100)
ax[2].set_title('Аргумент спектра', fontdict={'fontsize': 20})
plt.show()
```

Задача 3:

Самостоятельно ознакомившись с тем что такое Автокорреляционная функция рассчитать и отобразить графически Автокорреляционную функцию прямоугольного импульса: [np.zeros(8), np.ones(3), np.zeros(8)] , Напишите функцию вычисляющую автокорреляцию.

```
'''
   Автокорреляционная функция  показывает зависимость между сигналом и его 
   копией, сдвинутой по времени.

   Автокорреляционная функция помогает находить повторяющиеся участки во 
   временной последовательности, с её помощью можно находить несущую частоту 
   сигнала.

   Физический смысл АКФ это энергия сигнала. 

'''

# Так как автокорреляция это просто свертка сигнала с самим собой - напишем простую функцию:
def auto_corr(x):
    res = np.correlate(x, x, mode='same') # np.correlate - вычисление свертки
    return res 

# Сигнал
x = np.concatenate([np.zeros(8), np.ones(3), np.zeros(8)])

# Полная функция автокорреляции(ACF)
cfx = auto_corr(x)

xl = [x, cfx]

# Строим графики
plt.figure(figsize=(14, 5), dpi=80)
for i in range(2):
    plt.subplot(2, 1, i+1)
    plt.stem(xl[i], linefmt='C0', markerfmt='D', use_line_collection=True)
    plt.grid(True)
    plt.xlim([-0.5, x.size-0.5])
```
