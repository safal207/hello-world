Задача 1

Вероятность того, что стрелок поразит мишень, равна 0.7 

Составить закон распределения случайной величины  W – количества попаданий после 2 выстрелов.

```
# Имеем ТРИ исхода:
# A1 - стрелок 2 раза промахнулся
# A2 - стрелок попал оба раза
# A3 - стрелок попал один раз а один промахнулся

P_a1 = (1-0.7)*(1-0.7)
P_a2 = 0.7*0.7
print ("0:{}".format(P_a1))
print ("2:{}".format(P_a2))
# Вероятность A3 найдем через теорему о полной группе вероятностей
P_a3 = 1-(P_a1+P_a2)
print ("1:{}".format(P_a3))
W={0:P_a1, 1:P_a3, 2:P_a2}
print(W)
```

Задача 2

Мистер Хэ играет в европейскую рулетку по следующей системе: постоянно ставит 250 рублей на «красное». 
Составить закон распределения случайной величины  – его выигрыша. 

Вычислить математическое ожидание выигрыша и округлить его. 

Сколько в среднем проигрывает игрок с каждой ставки?

```
# Количество исходов на рулетке:
#37 = 36(18 красное и 18 черное) + 1 zero
```

```
import pandas as pd
```

```
# Мистер Хэ выигрывает 250 рублей в 18 случаях из 37, следовательно закон распределения его выигрыша имеет следующий вид:
w={250:18/37., -250:19/37.}
data = pd.DataFrame([w])
print(data)

# Вычислим математическое ожидание: M(x) = p1*x1 + p2*x2
mm = [ key * w[key] for key in w.keys()]
print("Мат ожидание игры: {} рубля".format(round(sum(mm),2)))
print("С каждой поставленной ставки игрок в среднем проигрывает: {} рубля".format(abs(round(sum(mm),2))))
```

Задача 4

Автобусы некоторого маршрута идут строго по расписанию и интервалом 5 минут. Составить функцию плотности случайной величины  X – времени ожидании очередного автобуса пассажиром, который наудачу подошёл к остановке. Найти вероятность того, что он будет ждать автобус не более 2x минут. 

```
# Дискретная или непрерывная?
# Время можно разделить на сколь угодно малый интервал - следовательно случайная величина непрерывная.
# Из условия нам известно что автобусы ходят СТРОГО по расписанию, следовательно в значении функции плотности КОНСТАНТА и мы имеем дело с РАВНОМЕРНЫМ распределением

# Попробуем вывести PDF
# 0 -  если х  от -inf до 0
# 1/5 - если х  от 0 до 5
# 0 если x > 5

# А теперь проверим корректность снова выведя PDF на этот раз через формулы и свойства плотности вероятности
#=========
# Запишем PDF
# 0 -  если х  от -inf до 0
# C - если х  от 0 до 5
# 0 если x > 5

# Теперь мы можем легко найти C по формуле для равномерного распределения C = 1/b-a , где b=5 и a=0
C = 1/(5.-0)
print (' Const = {}'.format(C))

# Перепишем PDF
# 0   - если х  от -inf до 0
# 0.2 - если х  от 0 до 5
# 0   - если x > 5

# Теперь вычислим вероятность прождать не более 2-х минут: (вероянтность на интервале 0,2)

from scipy.stats import uniform
p = uniform(loc=0,scale=5).cdf(2)   # Значения разбросаны на интервале от 0 до 5, это и указываем
print('Вероятность того, что он будет ждать автобус не более 2x минут: {}'.format ( p))
```

Задача 5

Рост мальчиков возрастной группы 15 лет есть нормально распределённая
случайная величина X с мат.ожиданием роста 161 см и среднеквадратичным отклонением 4 см. Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы. 


```
from scipy.stats import norm
p = norm (161, 4).cdf(158)- norm(161,4).cdf(152) # Считаем CDF для 158 и CDF для 152 
print('Доля костюмов для мальчиков, имеющих рост от 152 до 158 см: {}%'.format (round(100 * p, 3)))
```

Задача 7

Для задач 3 - 6 Вычислить среднее, дисперсию и 0.25 и 0.9 квантили

```
# Для решения этой задачи воспользуемся готовой библиотекой scipy
from scipy.stats import poisson, norm, uniform
dist_list = [] # Пустой лист куда запишется распределение с нужными параметрами

# Задаем распределения Задачи 3 и добавляем в dist_list
mu = 0
sigm = 4
norm_ = norm(mu,sigm)
dist_list.append(norm_)

# Задаем распределения Задачи 4 и добавляем в dist_list
unif = uniform(loc=0,scale=5)
dist_list.append(unif)

# Задаем распределения Задачи 5 и добавляем в dist_list
mu = 161
sigm = 4
norm_ = norm(mu,sigm)
dist_list.append(norm_)

# Задаем распределения Задачи 6 и добавляем в dist_list
mu = 100
std = 12
gaus1 = norm(mu, std)
dist_list.append(gaus1)


# Задаем функцию для расчета моментов распределения
def dist_moments(dist_):
  mean, var,_,_ = dist_.stats(moments='mvsk')
  print("Среднее: {}".format(mean))                    # Среднее
  print("Вариация: {}".format(var))                    # Дисперсия 
  print("Перцентиль 0.25: {}".format(dist_.ppf(0.25))) # Квантиль 25%
  print("Перцентиль 0.9: {}".format(dist_.ppf(0.9)))   # Квантиль 90%


# Теперь проходимся циклом по всем распределениям
for i,d in enumerate(dist_list):
  print("Для Задачи: {}".format(i+3))
  dist_moments(d)
  print('====================')
```
