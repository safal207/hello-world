Задача 6

Стрелок попадает в цель с одной и той же вероятностью при каждом выстреле. Какова эта вероятность, если вероятность хотя бы одного попадания при трех выстрелах равна 0,861.

```
# Пусть Р – вероятность попадания при каждом выстреле.
# Q – вероятность промаха при каждом выстреле.
# P = 1 - Q
# A1 – при 3 выстрелах стрелок попадёт в мишень хотя бы один раз;
# A2 – стрелок 3 раза промахнётся.
# P(A1) дана по условию задачи, тогда вероятность противоположного события: P(A2) =  1- P(A1)
# По теореме умножения вероятностей: P(A2) = QQQ = Q^3 = 1- P(A1)
# Отсюда искомая вероятность P = 1 - Q^1/3
# Таким образом все вместе:
1 - ((1-0.861)**(1/3.))
```

```
n = factorial(8)  # Варианты перестановок среди 8 человек 8!
# Для каждого варианта рассадки на 8 местах 2 человека могут рассесться двум способами:
# Один справа, второй слева или наоборот, для каждого такого варианта остальные 6 человек могут рассесться 6! способами, итого:
m = (8 * factorial(2)  * factorial(6)) 
print("вероятность того, что два определённых человека окажутся рядом: {}".format(m/float(n)))
```

На шахматную доску из 64 клеток ставят наудачу две ладьи, белого и чёрного цвета. С какой вероятностью они не будут «бить» друг друга?

```
n = factorial(64)/factorial(62) # Способами можно расставить m ладьей на n клетках: A(m,n) = n!/(m-n)! - формула размещений
# Проще вычислить случа когда ладьи БЬЮТ друг друга
# Это происходит если ладьи стоят на одной горизонтали или вертикали
# На каждой горизонтали из 8 клеток  кол-во сочетаний 2-х ладей можно вычислить по формуле размещений без повторений C(m,n) = n!/(n-m)!
m = factorial(8) / factorial(8-2) # способами можно разместить белую и чёрную ладью на 1й горизонтали.
# На доске 8 горизонталей и 8 вертикалей следовательно общее кол-во исходов когда ладьи  бьют друг друга:
m_o = (8*m)+(8*m)
m_f = n - m_o # кол-во исходов когда ладьи НЕ БЬЮТ друг друга
P = m_f/float(n)
print("c вероятностью {} ладьи не будут «бить» друг друга".format(P))
```

Вероятность поражения цели первым стрелком при одном выстреле равна 0,6. Вероятность того, что цель не поражена после выполнения первым и вторым стрелками по одному выстрелу равна 0,06. Какова вероятность поражения цели вторым стрелком при одном выстреле?

```
P_ab = 1 - 0.06 # Вероятность поражения после 2-х выстрелов 
#  P_ab = 0.6x+0.6(1-x) + 0.4*x - по формуле полной вероятности
# 0.6x+0.6-0.6x+0.4x=0.94
# 0.4x=0.94-0.6
x = (P_ab-0.6)/0.4

print("вероятность поражения цели вторым стрелком при одном выстреле: {}".format(x))
```
