```
#Установить terra_ai
!pip install terra_ai==0.0.0.34
#Импортировать модуль терра_ии
from terra_ai import терра_ии
```

```
# Загружаем базу трафик
терра_ии.загрузить_базу('ТРАФИК', справка=True)
```

```
# Создаем выборки для обучения нейронной сети
обучающий_генератор, тестовый_генератор, данные = терра_ии.создать_выборки_трафик()
```

```
# Создаем нейронную сеть

слои = 'Сверточный1D-1024 Выравнивающий Полносвязный-1024 Полносвязный-1'

# создание модели
нейронка = терра_ии.создать_сеть(слои = слои,
                                 задача = 'трафик')
# Обучаем модель
терра_ии.обучение_модели_трафик(нейронка,
    обучающий_генератор, тестовый_генератор,        
    количество_эпох = 40)

терра_ии.тест_модели_трафика(нейронка, данные)
```
