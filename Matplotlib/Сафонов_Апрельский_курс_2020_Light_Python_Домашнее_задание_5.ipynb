### **Задача 2**

Используя библиотеку Matplotlib, постройте графики функций $y = x^3$ и $y = x^2$ на одном графике. 

* Дайте название графику 
* Дайте названия осям
* Присвойте лейблы и выведите легенду
* Графики функций должны быть отрисованы пунктирными линиями (любыми из доступных на ваш вкус) 

```python
x = np.linspace(0, 10, 50) 
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]
```

```python

#Графики функций должны быть отрисованы пунктирными линиями (любыми из доступных на ваш вкус)
plt.plot(x, y1, 'b--',label='y=x^2') # Визуализируем график для значения y1 (и задаем ему название (легенду))
plt.plot(x, y2, 'g--',label='y=x^3') # Визуализируем график для значения y2 (и задаем ему название (легенду))

plt.title('Название y=x^2 и y=x^3') # Задаем название графика
plt.xlabel('ось X') # ось х
plt.ylabel('ось Y') # ось y
plt.legend(loc='upper left') # Вызываем отображение легенды на графике
plt.show() # Фиксируем график
```

### **Задача 5**

Сделайте resample данных по получасовому интервалу с использованием усреднения для всех признаков. Обратите внимание, что столбец с метками класса Occupancy вместо усреднения нужно взять "как есть", т.е. resample c методом "asfreq" или "bfill".

Постройте sns.jointplot между признаками Light и CO2

```python
ssr1 = df.Occupancy.resample('30min').bfill() #с метками класса Occupancy вместо усреднения нужно взять "как есть", т.е. resample c методом "asfreq" или "bfill"
ssr2 = df.iloc[:,:-1].resample('30min').mean() # Сделайте resample данных по получасовому интервалу с использованием усреднения для всех признаков
res_ssr = pd.concat([ssr1,ssr2])
```

```python
sns.jointplot(x=res_ssr.Light,y=res_ssr.CO2, data = res_ssr)
plt.show()
```
